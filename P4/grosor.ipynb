```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.abs(1 + 0.2063 * np.e**(-2j*1.52*np.pi/670*x))**2
d = np.linspace(0, 400, 5000)
fd = np.array([f(x) for x in d])
```

```python
(1+0.2063**2)/0.917
```

```python
plt.plot(d, fd)
plt.grid()
plt.grid(True, which="minor", axis="both", linestyle="dotted")
plt.minorticks_on()
plt.axvline(x=50, color="orange", linestyle="--", linewidth=1)
plt.axvline(x=150, color="orange", linestyle="--", linewidth=1)
plt.axhline(y=(1+0.2063**2)/0.917, xmin=0, xmax=0.26, linestyle="--", linewidth=1)
plt.axvline(x=93, ymin=0, ymax=0.61, linestyle="--", linewidth=1)
plt.text(95, 1.15, "(x=93nm, y=1.13[~])")
plt.xlabel("Grosor [nm]")
plt.ylabel("[~]")
plt.savefig("fotos/grosor.pdf", bbox_inches="tight")
```
