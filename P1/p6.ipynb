```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
```

```python
with open("powers/p6", "r") as f:
    x = []
    ysim = []
    for i, j in [tuple(i.split(" ")) for i in f.read().splitlines()]:
        x.append(float(j))
        ysim.append(float(i))

x = x[::-1]
ysim = ysim[::-1]
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, ysim, label="Valores simulados")
ax.plot(x, [np.sqrt(1.9)/np.sqrt(i) for i in x], label=r"Aprox. $\displaystyle\frac{A_1}{\sqrt{z}}$")
ax.plot(x, [1.9/i for i in x], label=r"Aprox. $\displaystyle\frac{A_2}{z}$")
ax.plot(x, [4/i**2 for i in x], label=r"Aprox. $\displaystyle\frac{A_3}{z^2}$")
ax.legend()
ax.set_xlabel("Distancia [mm]")
ax.set_ylabel("Potencia normalizada [W/W]")
ax.grid()
ax.minorticks_on()
ax.grid(True, which="minor", axis="y", linestyle="--")
ax.set_ylim(0.7995833333333333, 4.108750000000001)
plt.savefig("fotos/p6.pdf", bbox_inches="tight")
```
