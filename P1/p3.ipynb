```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def L_to_theta(L: float) -> float:
    return np.arcsin((0.6 * 5.5)/L) * 180/np.pi
```

```python
x = [i for i in range(16, 28 + 1) if i%2 == 0]

with open("powers/grating_aperture", "r") as f:
    y = [float(i) for i in f.read().splitlines()]

yteo = [ L_to_theta(i) for i in x ]
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, label="Valores simulados")
ax.plot(x, yteo, label="Valores teóricos")
ax.set_xlabel("Apertura [µm]")
ax.set_ylabel("Ángulo de máxima potencia [°]")
ax.legend()
ax.minorticks_on()
ax.grid()
ax.grid(True, which="minor", axis="y", linestyle="--")
plt.savefig("fotos/p3.pdf", bbox_inches="tight")
```
