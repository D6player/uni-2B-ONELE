```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def angle_to_mm(a: float) -> float:
    return np.tan(a * np.pi/180) * 10
```

```python
x = [2*i for i in range(1, 11)]

with open("powers/halfcentimeasures", "r") as f:
    yhalf = [angle_to_mm(float(i)) for i in f.read().splitlines()]

with open("powers/tenthcentimeasures", "r") as f:
    ytenth = [angle_to_mm(float(i)) for i in f.read().splitlines()]
```

```python
fig, ax1 = plt.subplots(figsize=(8, 4))
ax1.plot(x, yhalf, label="Apertura receptora de 0.5mm")
ax1.plot(x, ytenth, label="Apertura receptora de 0.1mm")
ax1.legend()
ax1.set_xlabel("Ancho de la apertura [m]")
ax1.set_ylabel("Semiancho del haz [mm]")
ax1.set_xlim(2, 20)
ax1.grid()
ax1.minorticks_on()
ax1.grid(True, which="minor", axis="y", linestyle="--")
ax2 = ax1.twinx()
ax2.set_ylabel("Semiangulo del haz []")
_ = ax2.set_ylim(1, 16)
plt.savefig("fotos/p1.pdf", bbox_inches="tight")
```
